```python

import folium
from folium import plugins
import branca

kashiwaeki = [35.86238453604905, 139.97089573914457]


m = folium.Map(location=kashiwaeki, zoom_start=15)

folium.CircleMarker(
    location=[35.86238453604905, 139.97089573914457],
    radius=280,
    color='#ffff00',
    fill_color='#ffff00'
).add_to(m)

html1="https://twitter.com/oudouyakashiwa"

group_1 = folium.FeatureGroup("その他").add_to(m)
group_2 = folium.FeatureGroup("二郎系").add_to(m)
group_3 = folium.FeatureGroup("中華そば").add_to(m)

iframe1 = branca.element.IFrame(html=html1, width=300, height=500)
popup1 = folium.Popup(iframe1, max_width=300)
oudouya=[35.86310705313056, 139.96749619209098]
folium.Marker(oudouya, popup=popup1, tooltip="王道家 柏店",icon=folium.Icon(color='blue')).add_to(group_1)

fukurou=[35.861254398368196, 139.9706365688336]
folium.Marker(fukurou, popup="二郎系　　", tooltip="ラーメン二九六",icon=folium.Icon(color='green')).add_to(group_2)

akebi=[35.86090086608687, 139.9741098153531]
folium.Marker(akebi, popup="中華そば　　　", tooltip="AKEBI",icon=folium.Icon(color='red')).add_to(group_3)

oremichi=[35.85572449449928, 139.97164485359403]
folium.Marker(oremichi, popup="二郎系　　", tooltip="俺の生きる道 柏店",icon=folium.Icon(color='green')).add_to(group_2)

homare=[35.86054457043861, 139.969043470783]
folium.Marker(homare, popup="背脂系　　", tooltip="こってりラーメン誉",icon=folium.Icon(color='blue')).add_to(group_1)

daikatsu=[35.854602772556795, 139.97357127917732]
folium.Marker(daikatsu, popup="中華そば(煮干し)　　　　　　　", tooltip="大勝",icon=folium.Icon(color='red')).add_to(group_3)

nishio=[35.86029125358195, 139.96834584466]
folium.Marker(nishio, popup="中華そば　　　", tooltip="麺処　にし尾",icon=folium.Icon(color='red')).add_to(group_3)

teraccyo=[35.86058639460725, 139.97286167078306]
folium.Marker(teraccyo, popup="博多ラーメン　　　　　", tooltip="三代目てらッちょ",icon=folium.Icon(color='blue')).add_to(group_1)

folium.LayerControl().add_to(m)

m
```
